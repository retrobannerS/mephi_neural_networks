```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA, FastICA
from sklearn.manifold import TSNE
from sklearn.neural_network import MLPRegressor
from mpl_toolkits.mplot3d import Axes3D
from sklearn.neural_network._base import ACTIVATIONS

```

```python
np.random.seed(42)
X, y = make_blobs(n_samples=300, centers=4, n_features=3,
                  cluster_std=1.2, random_state=42)
X = StandardScaler().fit_transform(X)

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –≤ 3D
fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c=y, cmap='viridis', s=30)
ax.set_title("–ò—Å—Ö–æ–¥–Ω—ã–µ 3D-–¥–∞–Ω–Ω—ã–µ")
ax.set_xlabel("X1"); ax.set_ylabel("X2"); ax.set_zlabel("X3")
plt.show()

```

```python
### 2. PCA (Principal Component Analysis)

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(6,5))
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, cmap='viridis', s=30)
plt.title("PCA")
plt.xlabel("PC1"); plt.ylabel("PC2")
plt.grid(True)
plt.show()

```

```python
### 3. ICA (Independent Component Analysis)
ica = FastICA(n_components=2, random_state=42)
X_ica = ica.fit_transform(X)

plt.figure(figsize=(6,5))
plt.scatter(X_ica[:,0], X_ica[:,1], c=y, cmap='viridis', s=30)
plt.title("ICA")
plt.xlabel("IC1"); plt.ylabel("IC2")
plt.grid(True)
plt.show()

```

```python
### 4. t‚ÄëSNE (non‚Äëlinear manifold learning)

tsne = TSNE(n_components=2, perplexity=30, random_state=42)
X_tsne = tsne.fit_transform(X)

plt.figure(figsize=(6,5))
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=y, cmap='viridis', s=30)
plt.title("t‚ÄëSNE")
plt.xlabel("Dim 1"); plt.ylabel("Dim 2")
plt.grid(True)
plt.show()

# UMAP –ø—Ä–æ–¥–æ–ª–∂–µ–Ω–∏–µ 
```

```python
### 5. –ê–≤—Ç–æ–∫–æ–¥–∏—Ä–æ–≤—â–∏–∫ (Autoencoder)

autoencoder = MLPRegressor(hidden_layer_sizes=(2,), activation='tanh',
                           max_iter=5000, random_state=42)
autoencoder.fit(X, X)  # –≤—Ö–æ–¥ = –≤—ã—Ö–æ–¥

# –§—É–Ω–∫—Ü–∏—è –¥–ª—è –∏–∑–≤–ª–µ—á–µ–Ω–∏—è —Å–∫—Ä—ã—Ç–æ–≥–æ –ø—Ä–µ–¥—Å—Ç–∞–≤–ª–µ–Ω–∏—è
def extract_hidden(model, X):
    act = ACTIVATIONS[model.activation]
    W, b = model.coefs_[0], model.intercepts_[0]
    return (np.dot(X, W) + b)

X_encoded = extract_hidden(autoencoder, X)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import FastICA
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from minisom import MiniSom

```

```python
digits = load_digits()
X = digits.data       # (n_samples, 64)
y = digits.target
X = StandardScaler().fit_transform(X)
```

```python
digits.data.shape
```

```python
plt.imshow(digits.data[323].reshape([8,8]), cmap="Grays")
```

```python
### 1. –ö–æ–Ω—Ç—Ä–æ–ª–∏—Ä—É–µ–º–æ–µ –æ–±—É—á–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: MLP –∏ —Å–∫—Ä—ã—Ç—ã–π —Å–ª–æ–π
# –û–±—É—á–∏–º –ø—Ä–æ—Å—Ç–æ–π –æ–¥–Ω–æ—Å–ª–æ–π–Ω—ã–π MLP-–∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ç–æ—Ä –∏ –∏–∑–≤–ª–µ—á—ë–º –≤—ã—Ö–æ–¥ —Å–∫—Ä—ã—Ç–æ–≥–æ —Å–ª–æ—è –∫–∞–∫ –ø—Ä–∏–∑–Ω–∞–∫–∏.

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=42,
                                                    stratify=y)

mlp = MLPClassifier(hidden_layer_sizes=(32,),
                    activation='tanh',
                    solver='adam',
                    max_iter=500,
                    random_state=42)
mlp.fit(X_train, y_train)

def extract_hidden_features(model, X):
    # –≤—Ö–æ–¥ X (n_samples, 64), –≤–µ—Å W0 (64, 32), bias b0 (32,)
    W0, b0 = model.coefs_[0], model.intercepts_[0]
    Z = X @ W0 + b0    # –ª–∏–Ω–µ–π–Ω–∞—è —á–∞—Å—Ç—å
    return np.tanh(Z)  

X_feats_mlp = extract_hidden_features(mlp, X_test)

# –í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–µ—Ä–≤—ã–µ –¥–≤–∞ –ø—Ä–∏–∑–Ω–∞–∫–∞ —Å–∫—Ä—ã—Ç–æ–≥–æ —Å–ª–æ—è
plt.figure(figsize=(6,5))
plt.scatter(X_feats_mlp[:,0], X_feats_mlp[:,1], c=y_test,
            cmap='tab10', s=30)
plt.title("–°–∫—Ä—ã—Ç—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ MLP (–ø–µ—Ä–≤—ã–µ 2 –Ω–µ–π—Ä–æ–Ω–∞)")
plt.xlabel("Neuron 1")
plt.ylabel("Neuron 2")
plt.grid(True)
plt.show()
```

```python
### 2. –ù–µ–∑–∞–≤–∏—Å–∏–º—ã–π –∞–Ω–∞–ª–∏–∑ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç (ICA)
# –ò–∑–≤–ª–µ–∫–∞–µ–º 10 –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã—Ö –∫–æ–º–ø–æ–Ω–µ–Ω—Ç–æ–≤ –±–µ–∑ —É—á—ë—Ç–∞ –º–µ—Ç–æ–∫.

ica = FastICA(n_components=2, random_state=111)
X_feats_ica = ica.fit_transform(X)

# –í–∏–∑—É–∞–ª–∏–∑–∏—Ä—É–µ–º –ø–µ—Ä–≤—ã–µ –¥–≤–µ IC
plt.figure(figsize=(6,5))
plt.scatter(X_feats_ica[:,0], X_feats_ica[:,1], c=y,
            cmap='tab10', s=30)
plt.title("ICA (2 –Ω–µ–∑–∞–≤–∏—Å–∏–º—ã–µ –∫–æ–º–ø–æ–Ω–µ–Ω—Ç—ã)")
plt.xlabel("IC 1")
plt.ylabel("IC 2")
plt.grid(True)
plt.show()
```

```python
### 3. –°–∞–º–æ–æ—Ä–≥–∞–Ω–∏–∑—É—é—â–∞—è—Å—è –∫–∞—Ä—Ç–∞ (SOM) –ö–æ—Ö–æ–Ω–µ–Ω–∞

# –ò–Ω–∏—Ü–∏–∞–ª–∏–∑–∞—Ü–∏—è SOM
som_size = 10
som = MiniSom(som_size, som_size, X.shape[1],
              sigma=1.0, learning_rate=0.5,
              )# random_seed=11)
som.train_random(X, 1000)  # 1000 –∏—Ç–µ—Ä–∞—Ü–∏–π

# –û—Ç–æ–±—Ä–∞–∑–∏–º, –∫—É–¥–∞ –ø–æ–ø–∞–¥–∞—é—Ç –ø—Ä–∏–º–µ—Ä—ã —Ä–∞–∑–Ω—ã—Ö –∫–ª–∞—Å—Å–æ–≤
plt.figure(figsize=(6,5))
for idx, x in enumerate(X[:200]):
    w = som.winner(x)  # –∫–æ–æ—Ä–¥–∏–Ω–∞—Ç–∞ –Ω–µ–π—Ä–æ–Ω–∞-–ø–æ–±–µ–¥–∏—Ç–µ–ª—è
    plt.text(w[0]+0.5, w[1]+0.5, str(y[idx]),
             color=plt.cm.tab10(y[idx]),
             fontdict={'size': 16})
plt.xlim([0, som_size])
plt.ylim([0, som_size])
plt.title("–ö–∞—Ä—Ç–∞ –∫–ª–∞—Å—Ç–µ—Ä–∏–∑–∞—Ü–∏–∏ SOM (–º–µ—Ç–∫–∏ –∫–ª–∞—Å—Å–æ–≤)")
plt.gca().invert_yaxis()
plt.show()
```

- **–ü–æ—à–∞–≥–æ–≤—ã–π —Å–ø—É—Å–∫ (Coordinate Descent):**

- **–ü–µ—Ä–µ–±–æ—Ä –ø–æ —Å–µ—Ç–∫–µ (Grid Search):**

¬†¬†–ü—Ä–µ–∏–º—É—â–µ—Å—Ç–≤–æ: –∑–∞—á–∞—Å—Ç—É—é —ç—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–µ–µ grid search –ø—Ä–∏ —É–≤–µ–ª–∏—á–µ–Ω–∏–∏ —á–∏—Å–ª–∞ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤.

```python
## 1. Coordinate Descent  

import numpy as np
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score

# –°–∏–Ω—Ç–µ—Ç–∏—á–µ—Å–∫–∏–π –¥–∞—Ç–∞—Å–µ—Ç
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=500, n_features=5, random_state=0)

# –ù–∞—á–∞–ª—å–Ω–∞—è –∫–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏—è –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
best_params = {'hidden_layer_sizes': (50,), 'alpha': 0.0001}
param_options = {
    'hidden_layer_sizes': [(10,), (50,), (100,)],
    'alpha': [1e-4, 1e-3, 1e-2]
}

# Coordinate Descent
for param in ['hidden_layer_sizes', 'alpha']:
    other = {k: v for k, v in best_params.items() if k != param}
    best_score = 0
    for val in param_options[param]:
        params = {**other, param: val}
        model = MLPClassifier(max_iter=3000, random_state=42, **params)
        score = cross_val_score(model, X, y, cv=3).mean()
        print(param, val, score)
        if score > best_score:
            best_score, best_params[param] = score, val

print("Best params (coordinate descent):", best_params)
```

```python


## 2. Grid Search  

from sklearn.model_selection import GridSearchCV

param_grid = {
    'hidden_layer_sizes': [(10,), (50,), (100,)],
    'alpha': [1e-4, 1e-3, 1e-2]
}
grid = GridSearchCV(
    estimator=MLPClassifier(max_iter=3000, random_state=0),
    param_grid=param_grid,
    cv=5,
    n_jobs=-1,           # –ø–∞—Ä–∞–ª–ª–µ–ª—å–Ω—ã–π –∑–∞–ø—É—Å–∫ –ø–æ –≤—Å–µ–º —è–¥—Ä–∞–º CPU
    verbose=2
)
grid.fit(X, y)
print("Grid Search best params:", grid.best_params_)
print("Best CV score:", grid.best_score_)
```

```python
## 3. Random Sampling  

from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import expon, randint

param_dist = {
    'hidden_layer_sizes': [(randint(10, 100).rvs(),) for _ in range(20)],
    'alpha': expon(scale=1e-3)         # —ç–∫—Å–ø–æ–Ω–µ–Ω—Ü–∏–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
}
rand_search = RandomizedSearchCV(
    estimator=MLPClassifier(max_iter=3000, random_state=0),
    param_distributions=param_dist,
    n_iter=20,                        # —á–∏—Å–ª–æ —Å–ª—É—á–∞–π–Ω—ã—Ö –ø—Ä–æ–±
    cv=5,
    random_state=0,
    n_jobs=-1,
    verbose=2
)
rand_search.fit(X, y)
print("Random Search best params:", rand_search.best_params_)
print("Best CV score:", rand_search.best_score_)
```
